```python
import matplotlib.pyplot as mp
import pandas as pd
import numpy as np
import h5py
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn import metrics
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Activation, BatchNormalization, Dropout, LeakyReLU
from tensorflow.keras.utils import to_categorical
from scipy.stats import zscore
from tensorflow.keras import optimizers
```

```python
dataSet = h5py.File("SVHN_single_grey1.h5")
```

```python
X_train = dataSet['X_train'][:]
X_test = dataSet['X_test'][:]
Y_train = dataSet['y_train'][:]
Y_test = dataSet['y_test'][:]
```

```python
X_train.shape
```

```python
fig, axes = mp.subplots(4,4)
for i, ax in enumerate(axes.flat): 
    if X_train[i].shape == (32, 32):
        ax.imshow(X_train[i])
    else:
        ax.imshow(X_train[i,:,:,0])
    ax.set_xticks([]); ax.set_yticks([])
    ax.set_title(Y_train[i])
```

From the above display grid, we can observe that the images are of single channel - a negative image alike. 

```python
dataSet.close()
```

Plotting the Distribution to understand how the data is spread:

```python
fig, (ax1, ax2) = mp.subplots(1, 2, sharex=True)

ax1.hist(Y_train, bins=10)
ax1.set_title("Training set")
ax1.set_xlim(1, 10)

ax2.hist(Y_test, color='g', bins=10)
ax2.set_title("Test set")

fig.tight_layout()
```

```python
#Normalizing the above data by Subtracting the mean of training set from individual point.
```

```python
X_train_mean = np.mean(X_train, axis = 0)
X_test_mean = np.mean(X_test, axis = 0)
```

```python
X_train_std = np.std(X_train, axis = 0)
X_test_std = np.std(X_test, axis = 0)
```

```python
X_train_normalized = (X_train-X_train_mean/X_train_std)
X_test_normalized = (X_test-X_test_mean/X_test_std)
```

Now splitting the X_train data into validation set and X_train values using train_test_split

```python
train_X, Val_X, train_Y, Val_Y = train_test_split(X_train_normalized, Y_train, random_state =1, test_size = .1)
```

```python
#Pre Processing the training X and Validate data that is to reduce the dimensions of the given data set from 3 to 2:
train_X = train_X.reshape((train_X.shape[0], -1))
test_X = X_test.reshape((X_test.shape[0], -1))
Val_X = Val_X.reshape((Val_X.shape[0],-1))
```

```python
print (train_X.shape,test_X.shape,Val_X.shape)
```

```python
#Converting the classes to binary classification using the one hot encoding process.
```

```python
train_Y = to_categorical(train_Y)
Y_test = to_categorical(Y_test)
Val_Y = to_categorical(Val_Y)
```

### The dataset is now divided as Training set, Validation Set and Testing set. 
Building a Deep Neural Network model around the given dataset:---

```python
#Creating a simple Feed forward model with activation function as signmoid.
```

```python
  model = Sequential()
  model.add(Dense(50, input_shape = (1024, )))
  model.add(Activation('relu'))
  model.add(Dense(50))
  model.add(Activation('relu'))
  model.add(Dense(50))
  model.add(Activation('relu'))
  model.add(Dense(50))
  model.add(Activation('relu'))
  model.add(Dense(10))
  model.add(Activation('softmax'))

sgd = optimizers.SGD(lr = 0.001)
model.compile(optimizer = sgd, loss ='categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(train_X,train_Y, epochs = 100, batch_size = 100, verbose =1)
```

```python
#Evaluating the model on actual test data
model.evaluate(test_X,Y_test)
```

```python
preds = model.predict_classes(test_X)
```

```python
preds
```

```python
preds_val = model.predict_classes(Val_X)
```

```python
preds_val
```

Thus, from above accuracy and loss metrices we can see that the model is poorly performing with accuracy of on 59.62% and a higher loss of 1.24. The observations are similar for both validation and the test data set. 

## Now building a Deep neural net with batch normalization, weight initializers and using RELU as an activation function.

```python
model1 = Sequential()
model1.add(Dense(100, input_shape = (1024, ), activation ='relu', kernel_initializer='he_normal'))
model1.add(BatchNormalization())
model1.add(Dense(300,activation = 'relu', kernel_initializer='he_normal'))
model1.add(BatchNormalization())
model1.add(Dense(500,activation = 'relu', kernel_initializer='he_normal'))
model1.add(BatchNormalization())
model1.add(Dense(300,activation = 'relu', kernel_initializer='he_normal'))
model1.add(BatchNormalization())
model1.add(Dense(10, activation = 'sigmoid', kernel_initializer='he_normal'))

model1.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics= ['accuracy'])
```

```python
model1.fit(train_X,train_Y, epochs = 100, batch_size = 100, verbose =1)
```

```python
model1.evaluate(Val_X, Val_Y)
```

Accuracy on the validation set is low that is 78.67% with loss of 1.263.

```python
model1.evaluate(test_X, Y_test)
```

Thus, unlike training accuracy, the model's accuracy is low on test data. Hence, model is overfitting.

## Now building a model with LeakyRelu as activation function and also including dropout in the model.

```python
model2 = Sequential()
model2.add(Dense(50, input_shape = (1024, ), kernel_initializer='he_normal'))
model2.add(BatchNormalization())
model2.add(LeakyReLU(alpha=0.01))
model2.add(Dropout(0.2))
model2.add(Dense(50, kernel_initializer='he_normal'))
model2.add(BatchNormalization())
model2.add(LeakyReLU(alpha=0.01))
model2.add(Dropout(0.2))
model2.add(Dense(100, kernel_initializer='he_normal'))
model2.add(BatchNormalization())
model2.add(LeakyReLU(alpha=0.01))
model2.add(Dropout(0.2))
model2.add(Dense(100, kernel_initializer='he_normal'))
model2.add(BatchNormalization())
model2.add(LeakyReLU(alpha=0.01))
model2.add(Dropout(0.2))
model2.add(Dense(10,activation = 'sigmoid'))
    
adam = optimizers.Adam()
model2.compile(optimizer = adam, loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model2.fit(train_X, train_Y, epochs = 200, batch_size = 100, verbose = 1)
```

```python
model2.evaluate(Val_X,Val_Y)
```

```python
model2.evaluate(test_X,Y_test)
```

```python
pred_val = model2.predict_classes(Val_X)
```

```python
pred_val
```

```python
pred_test = model2.predict_classes(test_X)
```
