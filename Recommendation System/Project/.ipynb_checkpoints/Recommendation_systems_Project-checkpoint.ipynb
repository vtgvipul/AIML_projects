**Now building a collaborative recommendation system**

From the train_dat creating matrix where rows will be userID and the columns will be productID with rating as its content.
This pivot matrix will be then used to create the SVD that is Singular Value Decomposition which will further be helpful to build the collborative recommender.

```python
train_dat.reset_index().head()
```

```python
train_dat.info()
```

```python
pivot_data = train_dat.pivot(index= 'userID', columns='productID', values='rating').fillna(0)
```

```python
pivot_data.head()
```

Now as per SVDs formula finding the U, Sigma and Vt. Later a dot product of (U and sigma) and Vt will be tabken to find the predicted values.

```python
U, sigma, Vt = svds(pivot_data, k = 25)
sigma = np.diag(sigma)
```

```python
pred_ratings_allUsers = np.dot(np.dot(U, sigma),Vt)
```

```python
pred_dataSet = pd.DataFrame(pred_ratings_allUsers, columns=pivot_data.columns)
```

```python
pred_dataSet.shape
```

```python
pred_dataSet['userID'] = pivot_data.index
```

```python
pred_dataSet.set_index(['userID'], inplace = True)
```

The above data frame is the predictions of ratings that all the users will do. Thus, if a user has never purchased a particular product, his liking and rating is predicted for the same and other items based on his ratings will be recommended.

```python
#Following function will take userID as an input and recommend the n number products for it based on above dataFrame

def recommendProduct(userID,n_recommendations):
    #user_index = userID-1
    user_ratings = pivot_data.loc[userID].sort_values(ascending = False)
    user_Preds = pred_dataSet.loc[userID].sort_values(ascending = False)
    user_recom_table = pd.concat([user_ratings,user_Preds], axis = 1)
    user_recom_table.index.name = 'Recommended Products'
    user_recom_table.columns = ['Actual_ratings', 'Predicted_ratings']
    user_recom_table = user_recom_table.loc[user_recom_table.Actual_ratings == 0]
    user_recom_table = user_recom_table.sort_values('Predicted_ratings', ascending = False)
    print("Following are the "+ str(n_recommendations) + ' for the user ' + str(userID) + ' :')
    return user_recom_table.head(n_recommendations)
```

```python
user_data_temp =recommendProduct(userID='A3CW0ZLUO5X2B1' , n_recommendations=37256)
```

Calculating the RMSE of the prediction model to gauge how reliable our recommnedations are.

```python
mean_pivot = pivot_data.mean()
mean_pred = pred_dataSet.mean()
```

```python
mean_pivot.head()
```

```python
mean_pred.head()
```

```python
rmse_data = pd.concat([mean_pivot,mean_pred],axis = 1)
rmse_data.columns = ['avg_ratings','avg_preds']
```

```python
RMSE = round((((rmse_data.avg_ratings - rmse_data.avg_preds) ** 2).mean() ** 0.5), 5)
```

Thus, from the above activity of building the popularity and collaborative based recommendation system, we can conclude and summarise the following:

1. Popularity based system is generalized and non personal. It will recommend the top N and popular product/ items in the database. No user preferences are taken into consideration.
    
    1.a. Data sparsity or density won't have much affect on these systems.
    
    
    
2. Collaborative filtering/ recommenders are user based systems that consider the user preferences by guaging the user's previous ratings and behaviour. The matrix based evaluation helps to predict the rating this user may give to a certain item, to which the user has never rated and on this basis, top N can be recommended to the user.
    
    2.a. Data density affects the recommendation. More sparse the data - vertically or horizantally will affect. If there are too many products and users, which are sparsely rated, such data will be not effective to predict ratings. For example for a single user, if we try to predict the ratings for Nth item out of a few 10 thousands of products, provided the user has only rated a handful of products. 
    
    2.b Collaborative filtering is helpful in cases where many users have equally rated throughout the oriduct spectrum.
    
    
3. Hybrid recommendations are more efficient where collaborative and content based systems can help to achieve a near actual prediction and recommendation in cases of sparse data.
