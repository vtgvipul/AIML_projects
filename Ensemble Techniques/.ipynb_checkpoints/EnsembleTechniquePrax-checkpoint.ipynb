```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as mp
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
dataSet = pd.read_csv("credit.csv")
dataSet.head(15)
```

```python
dataSet.shape
```

```python
dataSet.info()
```

```python
sb.pairplot(dataSet)
```

```python
dataSet.describe()
```

```python
for attr in dataSet.columns:
    if dataSet[attr].dtype == 'object':
        dataSet[attr] = pd.Categorical(dataSet[attr])
dataSet.head()
```

```python
dataSet.info()
```

```python
replaceStruct = {
                "checking_balance":     {"< 0 DM": 1, "1 - 200 DM": 2 ,"> 200 DM": 3 ,"unknown":-1},
                "credit_history": {"critical": 1, "poor":2 , "good": 3, "very good": 4,"perfect": 5},
                 "savings_balance": {"< 100 DM": 1, "100 - 500 DM":2 , "500 - 1000 DM": 3, "> 1000 DM": 4,"unknown": -1},
                 "employment_duration":     {"unemployed": 1, "< 1 year": 2 ,"1 - 4 years": 3 ,"4 - 7 years": 4 ,"> 7 years": 5},
                "phone":     {"no": 1, "yes": 2 },
                #"job":     {"unemployed": 1, "unskilled": 2, "skilled": 3, "management": 4 },
                "default":     {"no": 0, "yes": 1 } 
                    }
oneHotCols=["purpose","housing","other_credit","job"]
```

```python
dataSet = dataSet.replace(replaceStruct)
dataSet = pd.get_dummies(dataSet, oneHotCols)
```

```python
dataSet.head(20)
```

```python
xval = dataSet.drop('default', axis = 1)
yval = dataSet['default']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(xval, yval, test_size = .30, random_state = 1)
```

```python
decTree = DecisionTreeClassifier(criterion='gini', random_state=1)
decTree.fit(xtrain,ytrain)
```

```python
decTree.score(xtest,ytest)
```

```python
decTree.score(xtrain,ytrain)
```

```python
#Reducing the Overfitting of the decision tree:
decTree = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=1)
```

```python
decTree.fit(xtrain, ytrain)
```

```python
decTree.score(xtest,ytest)
```
